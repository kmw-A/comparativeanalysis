```python
import pandas as pd

# Load the data
data_path = '/Users/kevinfred/Downloads/emotion_data_f7cb2403-1c6b-4790-8cdc-4faee44e449c.csv'  # Replace with your actual data file path
data = pd.read_csv(data_path)
data

```

```python
# Convert all emotion columns to numeric, errors are coerced to NaNs
emotion_columns = ['Face API_neutral', 'Face API_happy', 'Face API_sad', 'Face API_angry', 'Face API_fearful', 'Face API_disgusted', 'Face API_surprised',
                   'Affectiva_joy', 'Affectiva_sadness', 'Affectiva_disgust', 'Affectiva_contempt', 'Affectiva_anger', 'Affectiva_fear', 'Affectiva_surprise',
                   'Affectiva_valence', 'Affectiva_engagement']
data[emotion_columns] = data[emotion_columns].apply(pd.to_numeric, errors='coerce').fillna(0)

```

```python
# Statistical Summary
stats_data = data[emotion_columns].describe()

stats_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


corr_matrix = stats_data.corr()

plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Heatmap of Emotion Correlations')
plt.show()
```
